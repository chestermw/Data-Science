```python
#Demonstration of LLN (Law of large numbers) using randint and randn
```

---

```python
import numpy as np
from numpy.random import randn
from numpy.random import randint
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#Create list of integers 0,1
xtrials = 1000
sets = 50
x2 = []

for i in range(sets):
    x1 = (list(randint(0,2,xtrials)))
    x2.append(list([round(x1.count(0)/xtrials,2)*100, round(x1.count(1)/xtrials,2)*100,50]))
    #x1 = list(randint(0,2,xtrials))
df = pd.DataFrame(x2)

plt.figure(figsize=(15,8))
sns.set_style('darkgrid')
plt.plot(df[0], c='Purple', label='zeroes')
plt.plot(df[1], c='Blue', label='ones')
plt.plot(df[2], c='Red', label='theoretical')
plt.ylim(0,100)
plt.legend(loc='best', prop={'size': 20})
plt.show()
```
